## Sorting Data

- sorting data= arranging data in a particular order, ascending or descending
- method used is sort_values()
-  has two paremeters, by and ascending
- by=column to sort by
- ascending=True or False
- used mostly when having numerical values


```python
#sortig data by fare in ascending order 
data.sort_values(by='Fare',ascending=True).head(10)
```

```python
#sorting data by fare in descending order
data=data.sort_values(by='Fare',ascending=False)
data.head(10)
```

## Feature engineering

- Feature engineering means creating new columns that helps in analysing  models(basically creting new columns that help in analysing)

```python
#feature engineering age column
data['Age_group']=pd.cut(data['Age'],bins=[0,12,19,59,100],labels=['Child','Teenager','Adult','Senior Adult'])
data.head(10)
```

```python
#extracting title from name column
data['Title']=data['Name'].apply(lambda name:name.split(',')[1].split('.')[0].strip())
data.head()
```

# merging the data frame
- similar to join in sql
- your dataset is split across multiple files or tables.

similarities between sets and dictionaries

```python
dict1={
    'id':[1,2,3,4],
    'name':['Brenda','Mutai','Leo','Mwenda']
}
```

```python
dict2={
    'id':[1,2,3,4],
    'fruits':['banana','mango','orange','apple']
}
```

```python
import pandas as pd
```

```python
data1=pd.DataFrame(dict1)
data1.head()
```

```python
data2=pd.DataFrame(dict2)
data2.head()
```

```python
#LOADING THE TWO DATAFRAMES
merged= pd.merge(data1,data2,on='id',how='left')
merged.head()
```
