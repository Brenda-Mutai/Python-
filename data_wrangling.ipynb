# Data Wrangling
- involves filtering and sorting data frames, feature engineering, grouping and aggregation.
- also merging and joining dataframes

```python
#loading titanic dataset
import pandas as pd
data=pd.read_csv(r'tested.csv')
data.head()
```

```python
#checking information about the dataset
data.info()
```

```python
#dropping cabin column
data=data.drop('Cabin',axis=1)
data.head()
```

```python
data.info()
```

```python
#filtering data --comparison operators--boolean indexing
#obtain adults column
adults=data[data['Age']>=18]
adults.head(11)
```

```python
#mean of age
data['Age'].mean()
```

```python
# adults mean age
adults['Age'].mean()
```

```python
#only males
male=data[data['Sex']=='male']
male.head()
```

```python
#data sex.value_counts()
data['Sex'].value_counts()
```

```python
#filtering male
male['Sex'].value_counts()
```

```python
#filtering rows with Q embarkment
em=data[data['Embarked']=='Q']
em.head()
```

- sorting data= arranging data in a particular order, ascending or descending
- method used is sort_values()
-  has two paremeters, by and ascending
- by=column to sort by
- ascending=True or False


```python
#sortig data by fare in ascending order
data.sort_values(by='Fare',ascending=True).head(10)
```

```python
#sorting data by fare in descending order
data=data.sort_values(by='Fare',ascending=False)
data.head(10)
```
