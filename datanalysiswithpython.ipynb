## File input/output
- use the >with> statement in file handling in python, to avoid forgeting to close the file.

```python
#writing to a file
with open('sample.txt', 'w') as file:
    file.write('Hello World!')
```

```python
#reading from a file
with open('sample.txt', 'r') as file:
    content = file.read()
    print(content)
```

- if your column has more than 50% missing then drop it, unless they are tied to another column
- in analysis ypu can mention that it had many missing data

```python
# renaming ratingd to ratings
df.rename(columns={'Ratingd':'Ratings'},inplace=True)
df.head()
```

```python
#checking for missing values
df.isnull().sum()
```

- filling the missing values
- fillna is used to fill  the missing values in columns


```python
#filling the missing values
#drop  the two columns
df=df.drop(columns=['Ratings','Review'],axis=1)
df.head()
```
